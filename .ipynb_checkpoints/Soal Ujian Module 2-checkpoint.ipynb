### 8. Missing Values
#### Berapa banyak NaN age data dan mean dari setiap who?
#### Isi data NaN dari kolom age dengan mean dari umur tergantung whonya dia apa (child/woman/man)
#### Kemudian print banyak NaN age data dan mean umur setiap who lagi.

```python
# looking for mean
GroupByWho = titanic.groupby('who')
GroupByWho.mean()

meanMan = GroupByWho.mean()['age']['man']
meanWoman = GroupByWho.mean()['age']['woman']
meanChild = GroupByWho.mean()['age']['child']

# Looking for Nan
nan_rows = titanic[titanic['age'].isnull()]
nan_rows.head()

NanMan = nan_rows.groupby('who').count()['alone']['man']
NanWoman = nan_rows.groupby('who').count()['alone']['woman']

# Output
print('Banyak Nan man & mean : ' + str(NanMan) + ' & ' + str(meanMan))
print('Banyak Nan woman & mean : ' + str(NanWoman) + ' & ' + str(meanWoman))
print('Banyak Nan chiled & mean : ' + '0' + ' & ' + str(meanChild))
```

```python
# making a function
def bulat(x):
   return round(x)

# calculating the mean of age for each 'who' group
NanAge = titanic.groupby('who')['age'].mean().apply(bulat)
NanAge

# filling the Nan with the mean of age
titanic['age'].fillna(titanic['who'].map(NanAge), inplace=True)
titanic['age'].head()

# rounding the value
titanic['age'] = titanic['age'].apply(bulat)
titanic.head()

# showing that there is no more NaN in each group that is why I put 0 directly for 'Banyak NaN' below
nan_rows1 = titanic[titanic['age'].isnull()]
nan_rows1.head()

# output
print('Banyak Nan man & mean : ' + '0' + ' & ' + str(meanMan)) # I put 0 directly for 'Banyak NaN'
print('Banyak Nan woman & mean : ' + '0' + ' & ' + str(meanWoman)) # I put 0 directly for 'Banyak NaN'
print('Banyak Nan chiled & mean : ' + '0' + ' & ' + str(meanChild)) # I put 0 directly for 'Banyak NaN'
```

 - __id__ = id table ini (PK)
 - __mean_fare__ = Mean Fare

 - __batas_min_fare__ = Mean Fare - (3 x Std Fare)
 - __batas_max_fare__ = Mean Fare + (3 x Std Fare)
 
#### Dan data2nya ada 3 (fare setiap class yaitu first, second, and third class)

```python
# I used dictionary of data to make the needed dataframe. Here is how I searched the value to be put in the dictionary.
# After I got the value, I put these value one by one to the dictionary

# looking for the value of mean_fare of each class

fare_mean1 = titanic[(titanic['class'] == 'Third')]
x1 = fare_mean1.mean()['fare']

fare_mean3 = titanic[(titanic['class'] == 'Second')]
x3 = fare_mean3.mean()['fare']

fare_mean2 = titanic[(titanic['class'] == 'First')]
x2 = fare_mean2.mean()['fare']

list_fareMean = [x1, x2, x3]
print('list of Fare_Mean:')
print(list_fareMean)

# looking for the value of fare_mean of each class
fare_std1 = titanic[(titanic['class'] == 'Third')]
xx1 = fare_mean1.std()['fare']

fare_std3 = titanic[(titanic['class'] == 'Second')]
xx3 = fare_mean3.std()['fare']

fare_std2 = titanic[(titanic['class'] == 'First')]
xx2 = fare_mean2.std()['fare']

list_fareStd = [xx1, xx2, xx3]
print('list of Fare_Std:')
print(list_fareStd)

# looking for the value of batas_min_fare of each class
xxx1 = x1 - (3*xx1)

xxx3 = x3 - (3*xx3)

xxx2 = x2 - (3*xx2)

list_batasMinFare = [xxx1, xxx2, xxx3]
print('list of Batas_Min_Fare:')
print(list_batasMinFare)

# looking for the value of batas_max_fare of each class
xxxx1 = x1 + (3*xx1)

xxxx3 = x3 + (3*xx3)

xxxx2 = x2 + (3*xx2)

list_batasMaxFare = [xxxx1, xxxx2, xxxx3]
print('list of Batas_Max_Fare:')
print(list_batasMaxFare)
```

#### Tambahkan column outlier (1 for yes and 0 for no)

```python
# Managing data of 'Third' class
titanicThird = titanic[titanic['class'] == 'Third']
titanicThird.head()

def outlierThird(item):
    if (item < -21.658875):
        return 1
    elif (item > 49.009975):
        return 1
    else:
        return 0

serThird = (titanic['fare']).apply(outlierThird)
serThird.head()

titanicThird.insert(16, "outlier", serThird)
titanicThird.head()

# Managing data of 'Second' class
titanicSecond = titanic[titanic['class'] == 'Second']
titanicSecond.head()

def outlierSecond(item):
    if (item < -19.590013):
        return 1
    elif (item > 60.914379):
        return 1
    else:
        return 0

serSecond = (titanic['fare']).apply(outlierSecond)
serSecond.head()

titanicSecond.insert(16, "outlier", serSecond)
titanicSecond.head()

# Managing data of 'first' class
titanicFirst = titanic[titanic['class'] == 'First']
titanicFirst.head()

def outlierFirst(item):
    if (item < -150.986430):
        return 1
    elif (item > 319.295805):
        return 1
    else:
        return 0

serFirst = (titanic['fare']).apply(outlierFirst)
serFirst.head()

titanicFirst.insert(16, "outlier", serFirst)
titanicFirst.head()
```
